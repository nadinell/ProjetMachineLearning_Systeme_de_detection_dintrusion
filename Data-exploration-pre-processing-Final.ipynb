## <font color='DarkSlateBlue'> A. Importance of IDS

In today's interconnected world, the widespread use of network technologies and the internet has led to an increased risk of cyber attacks and network protocol breaches. Addressing these challenges necessitates the implementation of robust Intrusion Detection Systems (IDS) to safeguard systems from malicious activities.

The primary objective of an Intrusion Detection System (IDS) is to fortify networks against various forms of cyber threats. Traditional firewalls, while effective against known attacks, fall short when dealing with novel and evolving threats. ## <font color='DarkSlateBlue'> B. Types of IDS

1.  **Signature-based Detection:** This method identifies specific patterns or signatures associated with known malicious activities. While effective against recognized threats, it might fail to detect new or modified attacks.
   
2. **Anomaly-based Detection:** Leveraging machine learning techniques, this approach compares real-time network traffic against historical data to identify deviations from normal behavior. While adept at detecting unknown threats, it can result in higher false negatives.
    - **Categories of IDS based on deployment:**
        - **Host-based IDS (HIDS):** Operates on individual devices, comparing current file states with previous ones to detect any alterations.
        - **Network-based IDS (NIDS):** Positioned within the network to analyze traffic and detect anomalies, alerting administrators to potential threats.
        - **Hybrid-based IDS:** Combination of HIDS and NIDS for comprehensive threat detection.

## <font color=' MidnightBlue'> Columns Descriptions:
###  <font color='MediumOrchid'> I. Basic Features of Network

1. **duration**: Time duration of the connection (Quantitative)
2. **protocol type**: Protocol used in connection (Categorical, 3 categories)
3. **service**: Destination network service used (Categorical, 70 categories)
4. **flag**: Status of the connection (e.g. REJ = connection rejected) (Categorical, 11 categories)
5. **src bytes**: Number of data bytes transferred from source to destination (Quantitative)
6. **dst bytes**: Number of data bytes transferred from destination to source (Quantitative)
7. **land**: Indicator whether port number and IP address of source and destination are equal. 1 if yes, otherwise 0 (Binary)
8. **wrong fragment**: Number of wrong fragments in connection (Quantitative)
9. **urgent**: Number of urgent packets (Quantitative)

###  <font color='MediumOrchid'> II. Connection Related Features

10. **hot**: Number of "hot" indicators in the content (Quantitative)
11. **num failed logins**: Number of failed login attempts (Quantitative)
12. **logged in**: 1 if successfully logged in, 0 otherwise (Binary)
13. **num compromised**: Number of "compromised" conditions (Quantitative)
14. **root shell**: 1 if root shell is obtained, 0 otherwise (Binary)
15. **su attempted**: 1 if "su root" command attempted or used, 0 otherwise (Quantitative; dataset contains value 2)
16. **num root**: Number of operations performed as a root or root accesses (Quantitative)
17. **num file creations**: Number of file creation operations (Quantitative)
18. **num shells**: Number of shell prompts (Quantitative)
19. **num access files**: Number of operations on access control files (Quantitative)
20. **num outbound cmds**: Number of outbound commands in an FTP session (Quantitative)
21. **is host login**: 1 if the login is from root or admin, 0 otherwise (Binary)
22. **is guest login**: 1 if the login is from guest, 0 otherwise (Binary)

###  <font color='MediumOrchid'> III. Time Related Traffic Features

23. **count**: Number of connections to the same destination host as the current connection in the past 2 seconds (Quantitative)
24. **srv count**: Number of connections to the same service (port number) as the current connection in the past 2 seconds (Quantitative)
25. **serror rate**: % of connections that have activated s0, s1, s2, or s3 flag among connections aggregated in count (Quantitative)
26. **srv serror rate**: % of connections that have activated s0, s1, s2, or s3 flag among connections aggregated in srv count (Quantitative)
27. **rerror rate**: % of connections that have activated REJ flag among connections aggregated in count (Quantitative)
28. **srv rerror rate**: % of connections that have activated REJ flag among connections aggregated in srv count (Quantitative)
29. **same srv rate**: % of connections to the same service among those aggregated in count (Quantitative)
30. **diff srv rate**: % of connections to the different service among those aggregated in count (Quantitative)
31. **srv diff host rate**: % of connections that were to different destination machines among the connections aggregated in srv count (Quantitative)

###  <font color='MediumOrchid'> IV. Host Based Traffic Features

32. **dst host count**: Count of the connections having the same destination IP address (Quantitative)
33. **dst host srv count**: Count of connections having the same port number (Quantitative)
34. **dst host same srv rate**: % of connections that were to different services among those in dst host count (Quantitative)
35. **dst host diff srv rate**: % of connections that were to different services among those in dst host count (Quantitative)
36. **dst host same src port rate**: % of connections that were to the same source port among those in dst host srv count (Quantitative)
37. **dst host srv diff host rate**: % of connections that were to different destination machines among those in dst host srv count (Quantitative)
38. **dst host serror rate**: % of connections that have activated the s0, s1, s2, or s3 flag among those in dst host count (Quantitative)
39. **dst host srv serror rate**: % of connections that have activated the s0, s1, s2, or s3 flag among those in dst host srv count (Quantitative)
40. **dst host rerror rate**: % of connections that have activated the REJ flag among those in dst host count (Quantitative)
41. **dst host srv rerror rate**: % of connections that have activated the REJ flag among those in dst host srv count (Quantitative)

### <font color='DarkSlateBlue'> Transforming the label values to the attack type instead of its subclasses

Mainly, we distinguish 4 attack types that are: </br>
    -DoS</br>
    -U2R</br>
    -R2L</br>
    -Probe</br>
Therefore, we aim for our label values to be either within these four classes, depending on the detected intrusion, either have the value normal in case of no detected attack. 

creating a method to replace the subclasses of each attack with the attack type within the label "class"

```python
def attack_type(attack_list, attack_type):
    train['class'].replace(attack_list, attack_type, inplace=True);
    test['class'].replace(attack_list, attack_type, inplace=True);
```

defining the list of subclasses of each attack type

```python
subclass_dos = ['apache2', 'back', 'land', 'neptune', 'mailbomb', 'pod', 'processtable', 'smurf', 'teardrop','udpstorm', 'worm']
subclass_u2r = ['buffer_overflow', 'loadmodule', 'perl', 'rootkit', 'ps', 'sqlattack', 'xterm']
subclass_r2l = ['ftp_write', 'guess_passwd', 'imap', 'multihop', 'phf', 'spy', 'warezclient', 'warezmaster', 'sendmail', 'named', 'snmpgetattack',
'snmpguess', 'xlock', 'xsnoop', 'httptunnel']
subclass_probe = ['ipsweep', 'nmap', 'portsweep', 'satan', 'mscan', 'saint']
```

applying the changes

```python
attack_type(subclass_dos, 'DoS')
attack_type(subclass_u2r, 'U2R')
attack_type(subclass_r2l, 'R2L')
attack_type(subclass_probe, 'Probe')
```

```python
def handle_outliers_iqr(data):
    column = ['src_bytes', 'dst_bytes']
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
```
